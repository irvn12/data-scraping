```python
import instaloader
from datetime import datetime

def download_info_post(post):
    tanggal_post = post.date.strftime("%Y-%m-%d")
    waktu_post = post.date.strftime("%H:%M:%S")
    jumlah_likes = post.likes

    return tanggal_post, waktu_post, jumlah_likes

# Meminta input username dan password dari pengguna
username = input("Masukkan nama pengguna Instagram: ")
password = input("Masukkan kata sandi Instagram: ")
cari_pengguna = input("Username target: ")

L = instaloader.Instaloader()

# Coba untuk login
try:
    L.login(username, password)
except instaloader.exceptions.InstaloaderException as e:
    print(f"Error during login: {e}")
    exit()

try:
    profil = instaloader.Profile.from_username(L.context, cari_pengguna)
except instaloader.exceptions.ProfileNotExistsException:
    print(f"Profil dengan username '{cari_pengguna}' tidak ditemukan.")
    exit()

# Mengakses informasi profil pengguna
print("Informasi Profil untuk", cari_pengguna)
print("Username:", profil.username)
print("Nama Lengkap:", profil.full_name)
print("Bio:", profil.biography)
print("Pengikut:", profil.followers)
print("Mengikuti:", profil.followees)
print("Postingan:", profil.mediacount)
print()

# Menampilkan data postingan
for post in profil.get_posts():
    tanggal_post, waktu_post, jumlah_likes = download_info_post(post)

    print("Tanggal Post:", tanggal_post)
    print("Waktu Post:", waktu_post)
    print("Jumlah Likes:", jumlah_likes)
    print()
```

```python
import openpyxl
# Membuat file Excel
file_excel = "data_instagram.xlsx"
buku_kerja = openpyxl.Workbook()
lembar = buku_kerja.active

# Menambahkan header kolom
lembar["A1"] = "Tanggal Post"
lembar["B1"] = "Waktu Post"
lembar["C1"] = "Jumlah Likes"

baris = 2
for post in profil.get_posts():
    tanggal_post, waktu_post, jumlah_likes = download_info_post(post)

    lembar[f"A{baris}"] = tanggal_post
    lembar[f"B{baris}"] = waktu_post
    lembar[f"C{baris}"] = jumlah_likes

    baris += 1

# Menyimpan file Excel
buku_kerja.save(file_excel)
print("Data berhasil disimpan dalam file", file_excel)
```

```python
import pandas as pd

# Step 1: Read data from source Excel files
folder_path = 'D:\\MSIB Mari Belajar\\Project Capstone\\'
source1_file_path = folder_path + 'instagram_data.xlsx'
source2_file_path = folder_path + 'instagram_data_fix.xlsx'
source3_file_path = folder_path + 'data_instagram.xlsx'

# Read the correct file path for source1_data
source1_data = pd.read_excel(source1_file_path)
source2_data = pd.read_excel(source2_file_path)
source3_data = pd.read_excel(source3_file_path)

# Display columns in each dataframe to identify the common column
print("Columns in source1_data:", source1_data.columns)
print("Columns in source2_data:", source2_data.columns)
print("Columns in source3_data:", source3_data.columns)

# Identify the common column for merging
common_column = 'Username'  # Modify this based on the actual common column

# Step 2: Merge dataframes based on the common column
try:
    target_data = source1_data.merge(source2_data[[common_column, 'Komentar', 'Tanggal']], on=common_column, how='left')
    target_data = target_data.merge(source3_data[[common_column, 'Nama Lengkap', 'Bio', 'Pengikut', 'Mengikuti', 'Postingan', 'Tanggal Post', 'Waktu Post', 'Jumlah Likes', 'Komentar']], on=common_column, how='left')
except KeyError as e:
    print(f"KeyError: {e}")
    print("Ensure that the common column exists in the specified source dataframe.")

# Step 3: Save the merged dataframe to a new Excel file
output_file_path = folder_path + 'instagram_data_real_fix.xlsx'
target_data.to_excel(output_file_path, index=False)

print(f"Data berhasil disimpan dalam file {output_file_path}")
```

```python
import os
import json
import pandas as pd

folder_path = r'D:\MSIB Mari Belajar\Project Capstone\maribelajar_id'
output_file = r'D:\MSIB Mari Belajar\Project Capstone\maribelajar_data.xlsx'

# List all JSON files in the folder
json_files = [file for file in os.listdir(folder_path) if file.endswith('.json')]

# Create an empty list to store the extracted data
data = []

for json_file in json_files:
    file_path = os.path.join(folder_path, json_file)
    
    # Read the JSON file
    with open(file_path, 'r') as f:
        data_list = json.load(f)
    
    # Iterate over each object in the list
    for item in data_list:
        # Extract username and time from the JSON data
        username = item['owner']['username']
        time = item['created_at']
        
        # Append the data to the list
        data.append({'File': json_file, 'Username': username, 'Time': time})

# Create a DataFrame from the list of data
df = pd.DataFrame(data)

# Save the DataFrame to Excel
df.to_excel(output_file, index=False)
```
