```python
# Inisialisasi driver
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.linkedin.com/company/maribelajar/")

# Menunggu hingga elemen muncul dan terlihat
wait = WebDriverWait(driver, 30)

# Ambil informasi perusahaan
company_name_element = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="main-content"]/section[1]/section/div/div[2]/div[1]')))
company_name = company_name_element.text.strip()

company_description_element = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="main-content"]/section[1]/section/div/div[2]/div[1]/h2')))
company_description = company_description_element.text.strip()

website_element = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="main-content"]/section[1]/div/section[1]/div/dl/div[1]/dd')))
website = website_element.text.strip()

industry_element = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="main-content"]/section[1]/div/section[1]/div/dl/div[2]/dd')))
industry = industry_element.text.strip()

company_size_element = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="main-content"]/section[1]/div/section[1]/div/dl/div[3]/dd')))
company_size = company_size_element.text.strip()

headquarters_element = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="main-content"]/section[1]/div/section[1]/div/dl/div[4]/dd')))
headquarters = headquarters_element.text.strip()

company_type_element = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="main-content"]/section[1]/div/section[1]/div/dl/div[5]/dd')))
company_type = company_type_element.text.strip()

founded_element = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="main-content"]/section[1]/div/section[1]/div/dl/div[6]/dd')))
founded = founded_element.text.strip()

specialties_element = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="main-content"]/section[1]/div/section[1]/div/dl/div[7]/dd')))
specialties = specialties_element.text.strip()

locations_element = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="address-0"]')))
locations = locations_element.text.strip()

print("Company Name:", company_name)
print("Company Description:", company_description)
print("Website:", website)
print("Industry:", industry)
print("Company Size:", company_size)
print("Headquarters:", headquarters)
print("Company Type:", company_type)
print("Founded:", founded)
print("Specialties:", specialties)
print("Locations:", locations)

# Menutup browser
driver.quit()
```

```python
# Inisialisasi driver
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.linkedin.com/feed/hashtag/?keywords=maribelajar")

# Scroll ke bawah untuk memuat lebih banyak postingan
SCROLL_PAUSE_TIME = 2
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# Menemukan semua elemen berdasarkan XPath yang diberikan
elements = driver.find_elements(By.XPATH, '/html/body/div[5]/div[3]/div/div[2]/div/div[2]/main/div[2]/div/div[2]/div[3]/div/div')

# Mengambil tanggal dan waktu postingan
tanggal_waktu = elements[0].find_element(By.XPATH, './/time').get_attribute("datetime")
print("Tanggal dan Waktu:", tanggal_waktu)

# Mengambil jumlah like
elemen_like = elements[0].find_element(By.XPATH, './/span[@aria-label="Like"]/span')
jumlah_like = elemen_like.get_attribute("aria-label")
print("Jumlah Like:", jumlah_like)

# Mengambil jumlah komentar
elemen_komentar = elements[0].find_element(By.XPATH, './/span[@aria-label="Comment"]/span')
jumlah_komentar = elemen_komentar.get_attribute("aria-label")
print("Jumlah Komentar:", jumlah_komentar)

# # Menutup browser
# driver.quit()
```

```python
import json
from openpyxl import Workbook

# Buka file JSON dengan encoding 'utf-8'
with open('D:/MSIB Mari Belajar/Project Capstone/Linkedin maribelajar data/dataset_linkedin-post-search-scraper_2023-12-04_19-36-14-180.json', encoding='utf-8') as file:
    # Baca konten JSON
    data = json.load(file)

# Membuat workbook baru
wb = Workbook()
ws = wb.active

# Menambahkan header
header = ["Text", "Title", "Time Since Posted", "Likes Count", "Comments Count", "Author Profile ID", "Author Headline", "Author Full Name"]
ws.append(header)

# Menambahkan data JSON ke dalam worksheet
for post in data:
    text = post.get("text", "")
    title = post.get("title", "")
    time_since_posted = post.get("timeSincePosted", "")
    likes_count = post.get("likesCount", "")
    comments_count = post.get("commentsCount", "")
    author_profile_id = post.get("authorProfileId", "")
    author_headline = post.get("authorHeadline", "")
    author_full_name = post.get("authorFullName", "")

    # Menambahkan data ke dalam baris baru pada worksheet
    row = [text, title, time_since_posted, likes_count, comments_count, author_profile_id, author_headline, author_full_name]
    ws.append(row)

# Menyimpan workbook ke dalam file Excel
output_file = 'output.xlsx'
wb.save(output_file)

print("Data berhasil disimpan ke dalam file Excel:", output_file)
```
